```python
# Importing libraries

import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv")
data
```

```python
X = data.drop(['PrecipitationSumInches'], axis = 1)
```

```python
Y = data["PrecipitationSumInches"]
```

```python
# Reshaping it into 2D vector

Y = Y.values.reshape(-1, 1)
```

```python
day_index = 798
days = [i for i in range(Y.size)]
```

```python
# Initialise the linear regression classifier

clf = LinearRegression()

# Train the classifier

clf.fit(X, Y)
```

```python
# Plot a graph

print("The Precipitation trend graph: ")
plt.scatter(days, Y, color = 'g')
plt.scatter(days[day_index], Y[day_index], color = 'r')
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = X.filter(['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH'])
```

```python
print("The Precipitation trend graph:")

for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i+1)
    # Green scatter for all days
    plt.scatter(days, x_vis.iloc[:, i], color='g')
    # Red scatter for a specific day (highlight)
    plt.scatter(days[day_index], x_vis.iloc[day_index, i], color='r')
    plt.title(x_vis.columns[i])

plt.tight_layout()
plt.show()
```
